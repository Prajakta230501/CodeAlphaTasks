## Introduction:
Humans process visuals faster than text, making data visualization a powerful tool for turning raw data into meaningful insights. Well-designed visuals not only enhance understanding but also tell compelling stories that drive action.

In this task, we'll visualize book data scraped from a website to uncover insights like price distribution, rating counts, and more.

---


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set seaborn style
sns.set(style="whitegrid")
```

## Load Dataset

We are using a CSV file named `books.csv`, which contains book titles, prices, and ratings scraped from books.toscrape.com.


```python
# Load dataset
df = pd.read_csv("books.csv")
df.head()
```

## Visualize Book Prices

Letâ€™s check how book prices are distributed.


```python
plt.figure(figsize=(10, 6))
sns.histplot(df["Price "], bins=20, kde=True, color='orange')
plt.title("Distribution of Book Prices", fontsize=14)
plt.xlabel("Price")
plt.ylabel("Count")
plt.show()
```

## Book Ratings Count

Letâ€™s count how many books fall under each rating.


```python
plt.figure(figsize=(8, 5))
sns.countplot(x="Rating", data=df, palette="coolwarm")
plt.title("Number of Books by Rating", fontsize=14)
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

## Price vs Rating

Letâ€™s see if there is a trend between price and rating.


```python
plt.figure(figsize=(10, 6))
sns.scatterplot(x="Rating", y="Price ", data=df, hue="Rating", palette="Set2")
plt.title("Book Price vs Rating", fontsize=14)
plt.xlabel("Rating")
plt.ylabel("Price ")
plt.show()
```

## Conclusion

- Most books are priced between 30 and 60 units.
- Rating 4 is the most common among books.
- There isnâ€™t a strong correlation between price and rating.
